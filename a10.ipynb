```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

# Create x values
x = np.linspace(-5, 5, 100)

# Calculate y values
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = []
for i in x:
    if i<0:
        y_relu.append(0)
    else:
        y_relu.append(i)

print("Activation Functions Created!")
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y_sigmoid, label='Sigmoid', color='blue')

# Sigmoid
plt.title('Sigmoid')
plt.grid(True)

plt.show()
# Tanh

plt.plot(x, y_tanh, 'r-')
plt.title('Tanh')
plt.grid(True)

plt.show()
# ReLU

plt.plot(x, y_relu, 'g-')
plt.title('ReLU')
plt.grid(True)

plt.show()
```
